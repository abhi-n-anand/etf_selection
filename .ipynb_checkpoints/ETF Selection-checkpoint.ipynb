```python
import pandas as pd
import numpy as np 
import matplotlib.pyplot as plt 
import yfinance as yf 

%matplotlib inline
```

```python
ETFS_AND_MFS = [
"QQQ",
"QQQM",
"VTI",
"VOO",
"SPY",
"VTSAX",
"SQQQ",
"TQQQ"
]

query_str = " ".join(ETFS_AND_MFS)
query_str
```

```python
data = yf.download(query_str, period="max")
```

```python
data.index = data.index.to_period('D')
```

```python
data['Adj Close']['2020':].plot(title="Absolute Value of ETF and MF Instruments From 2020")
plt.legend(bbox_to_anchor=(1, 1))
```

```python
var = data['Adj Close']['2020':].var(axis=0).sort_values().astype(int)
var
```

```python
mean = data['Adj Close']['2020':].mean(axis=0).sort_values().astype(int)
mean
```

```python
data['Adj Close'].iloc[-1].astype(int).loc[mean.index]
```

```python
((data['Adj Close'].iloc[-1].astype(int).loc[mean.index] - mean) / mean).round(2) # Percentage of overweightness
```
